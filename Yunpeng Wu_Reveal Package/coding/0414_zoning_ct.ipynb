# Spatially join CTs to PLUTO data

Spatial join: https://geopandas.org/docs/user_guide/mergingdata.html

```python
pluto09 = pluto09.to_crs("EPSG:4326")
ct_pluto09 = gpd.sjoin(gdf, pluto09, op='contains')
```
